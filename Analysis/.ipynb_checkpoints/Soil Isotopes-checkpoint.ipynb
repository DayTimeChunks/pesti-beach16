```python
import datetime
import os

import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import *
import plotly.graph_objs as go

import numpy as np

from obs_data import *

print(os.getcwd())
```

```python
init_notebook_mode(connected=True)
```

## Observed

```python
n_obs = getSoilData("North", tss=True)
v_obs = getSoilData("Valley", tss=True)
s_obs = getSoilData("South", tss=True)
# s_obs.head()

nConc = n_obs['Conc.mug.g.dry.soil'].tolist()
vConc = v_obs['Conc.mug.g.dry.soil'].tolist()
sConc = s_obs['Conc.mug.g.dry.soil'].tolist()
```

## Simulated

```python
# Model version and mc-run number:
version = "v9"
folder = "1"

# Working computer
PC = False
if PC:
    back = "\\"  
    sim_path = version + back  
else:
    back = "/"
    sim_path = version + back
    
print(sim_path)
```

```python
col = 'Conc.North'
c_north_sim = pd.read_table(sim_path + folder + back + "resM_norCONC.tss", skiprows=4, delim_whitespace=True,
                         names=['Jdays', col],
                         header=None)

simN = c_north_sim['Conc.North'].tolist()
```

```python
col = 'Conc.Valley'
c_val_sim = pd.read_table(sim_path + folder + back + "resM_valCONC.tss", skiprows=4, delim_whitespace=True,
                         names=['Jdays', col],
                         header=None)

simV = c_val_sim['Conc.Valley'].tolist()
```

```python
col = 'Conc.South'
c_sou_sim = pd.read_table(sim_path + folder + back + "resM_souCONC.tss", skiprows=4, delim_whitespace=True,
                         names=['Jdays', col],
                         header=None)

simS = c_sou_sim['Conc.South'].tolist()
```

## Plotting

```python
# Julian days
x = s_obs.Jdays.tolist()

start = 169
end = 300
```

### Observed

```python
nObs = go.Scatter(
    x = x[start:end],
    y = nConc[start:end],
    mode = 'markers',
    name = 'North Conc.',
    marker=dict(
        color= 'red', # 'rgba(204, 204, 204, 0.95)',
        line=dict(
            color= 'black', #'rgba(217, 217, 217, 1.0)',
            width=1,
        ),
        symbol='circle',
        size=10)
)

vObs = go.Scatter(
    x = x[start:end],
    y = vConc[start:end],
    mode = 'markers',
    name = 'Valley Conc.',
    marker=dict(
        color= 'green', # 'rgba(204, 204, 204, 0.95)',
        line=dict(
            color= 'black', #'rgba(217, 217, 217, 1.0)',
            width=1,
        ),
        symbol='circle',
        size=10)
)

sObs = go.Scatter(
    x = x[start:end],
    y = sConc[start:end],
    mode = 'markers',
    name = 'South Conc.',
    marker=dict(
        color= 'blue', # 'rgba(204, 204, 204, 0.95)',
        line=dict(
            color= 'black', #'rgba(217, 217, 217, 1.0)',
            width=1,
        ),
        symbol='circle',
        size=10)
)
```

### Simulated

```python
traceN = go.Scatter(
    x = x[start:end],
    y = simN[start:end],
    mode = 'lines',
    name = 'North Sim.',
    marker=dict(
        color= 'red')
)

traceV = go.Scatter(
    x = x[start:end],
    y = simV[start:end],
    mode = 'lines',
    name = 'Valley Sim.',
    marker=dict(
        color='green')
)

traceS = go.Scatter(
    x = x[start:end],
    y = simS[start:end],
    mode = 'lines',
    name = 'South Sim.',
    marker=dict(
        color= 'blue')
)
```

```python
data = [nObs, vObs, sObs, traceN, traceV, traceS]
```

```python
%matplotlib inline
import plotlycred

online = True
if online:
    # fig = py.iplot(data) 
    fig = go.Figure(data=data) # , layout=layout)
    py.image.save_as(fig, filename='SoilsConc.png')
else:
    iplot(data)
```
