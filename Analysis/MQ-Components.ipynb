```python
import datetime
import os

import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import *
import plotly.graph_objs as go
init_notebook_mode(connected=True)

import numpy as np

from obs_data import *

print(os.getcwd())
```

```python
PC = True
analysis = False
short_sim = True
version = "test_runoff"
folder = "1"
timestamp = 'julian'
```

```python
if analysis:
    dot = ''
else:
    dot = '..\\'
    
if PC:
    back = "\\"  
    sim_path = dot + version + back  
else:
    back = "/"
    sim_path = version + back
    
print(sim_path)
```

## Observed

```python
var = 'CumMass_g'
obs = getOutletData(var) # grams
```

```python
if short_sim:
    s_obs = 166 - 1
    e_obs = 300
else:
    # Julian days
    s_obs = 0
    e_obs = 300

# Calendar days
if timestamp == 'julian':
    # Julian days
    x = obs.Jdays.tolist()[s_obs:e_obs]
else:
    # Calendar days
    dates = getTime(sep=";")
    x = dates.Date.tolist()[s_obs:e_obs]
    # dates = getTime(sep=";")
    # x = dates.Date.tolist()[s:e]
```

### Import

```python
# Simulated Outlet
sims = ['cumEXP_Smet_g', 'cumLF_L', 'cumADR_L', 'cumROz0_L']
names = ['EXP Sim. (Smet g)', 'LatFlux Sim. (Smet g)', 'ADR Sim. (Smet g)', 'Roff Sim. (Smet g)']
palette = ['blue', 'purple', 'rgb(178,223,138)', '#F44336', '#3F51B5', 'orange']

data = []  # <- store all graphing data

mObs = obs['CumMass_g'].tolist()[s_obs:e_obs]
traceObs = go.Scatter(
    x = x,
    y = mObs,
    mode = 'lines',
    name = 'Cum. Exp (S-met g)',
    marker=dict(
        color= 'green')
)
data.append(traceObs)

col = 'sim'
for i in range(len(sims)):    
    filename = "resM_" + sims[i] + ".tss"
    sim_mass = pd.read_table(sim_path + folder + back + filename, 
                                skiprows=4, delim_whitespace=True,
                                names=['Jdays', col],
                                header=None)
    mSim = sim_mass['sim'].tolist()
    trace = go.Scatter(
        x = x,
        y = mSim,
        mode = 'lines',
        name = names[i],
        marker = dict(
            color= palette[i])
    )
    data.append(trace)
```
