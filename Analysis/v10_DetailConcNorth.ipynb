## Observed

### Detailed

```python
det_name = "ug.g"
detailed = ["n1", "n2", "n3", "n4", "n5", "n7", "n8"]
pallete = ["#F44336", '#673AB7', '#3F51B5', '#009688', '#FF9800', '#4CAF50', '#03A9F4']
# red, purple, indigo, teal-green, orange, green, light-blue

obs_list = []
for i in range(len(detailed)):
    name = detailed[i] + ".tss"
    o = getDetailed(name)
    o = o[det_name].tolist()
    obs_list.append(o)
```

## Simulated

```python
# Model version and mc-run number:
version = "v10"
folder = "1"

# Working computer
PC = False
if PC:
    back = "\\"  
    sim_path = version + back  
else:
    back = "/"
    sim_path = version + back
    
print(sim_path)
```

```python
# Time series name
col = 'sim'
sim_list = []
for i in range(len(detailed)):
    name_tss = detailed[i] + 'CONC'
    filename = "resM_" + name_tss + ".tss"
    sim = pd.read_table(sim_path + folder + back + filename, 
                                skiprows=4, delim_whitespace=True,
                                names=['Jdays', col],
                                header=None)
    sim_list.append(sim[col].tolist())
```

## Plotting

```python
# Julian days
# x = s_obs.Jdays.tolist()
dates = getTime(sep=";")
x = dates.Date.tolist()

start = 175
end = 300
```

### Observed (markers)

```python
# Add markers to data
data = []
for i in range(len(detailed)):
    
    
    obs = go.Scatter(
        x = x[start:end],
        y = obs_list[i][start:end],
        mode = 'markers',
        name = detailed[i],
        marker=dict(
            color= pallete[i],
            line=dict(
                color= 'black', #'rgba(217, 217, 217, 1.0)',
                width=1,
            ),
            symbol='star',
            size=10)
    )
    data.append(obs)
```

### Simulated

```python
# Add sim lines to data
for i in range(len(detailed)):
    if detailed[i][0] == "n":
        color = 'red'
    elif detailed[i][0] == "t":
        color = 'green'
    else:
        color = 'blue'
    pobj = go.Scatter(
        x = x[start:end],
        y = sim_list[i][start:end],
        mode = 'lines',
        name = detailed[i] + " Sim.",
        marker=dict(color=pallete[i]))
    data.append(pobj)
```

```python
%matplotlib inline
import plotlycred
figname = 'NorthConcD3.png'

online = False
if online:
    # fig = py.iplot(data) 
    fig = go.Figure(data=data) # , layout=layout)
    py.image.save_as(fig, filename=figname)
else:
    iplot(data)
```
