```python
import datetime
import os

import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import *
import plotly.graph_objs as go
init_notebook_mode(connected=True)

import numpy as np

from obs_data import *

print(os.getcwd())
```

```python
PC = False
version = "v10"
folder = "1"
```

```python
if PC:
    back = "\\"  
    sim_path = version + back  
else:
    back = "/"
    sim_path = version + back
    
print(sim_path)
```

### Water balance error

```python
# Q error
col = "sim"
name_tss = "global_waterMB"
filename = "resW_" + name_tss + ".tss"

wb_m3 = pd.read_table(sim_path + folder + back + filename, 
                            skiprows=4, delim_whitespace=True,
                            names=['Jdays', col],
                            header=None)

real = wb_m3.sim.cumsum()

absol = wb_m3.sim.abs().cumsum()
```

```python
# Julian days
s = 180
e = 290
```

```python
dates = getTime(sep=";")
x = dates.Date.tolist()[s:e]
wb = wb_m3.tolist()[s:e]
re = real.tolist()[s:e]
ab = absol.tolist()[s:e]
```

```python
traceWB = go.Scatter(
    x = x,
    y = wb,
    mode = 'lines',
    name = 'Catchment Error (m3)',
    marker=dict(
        color= 'green')
)

traceRE = go.Scatter(
    x = x,
    y = re,
    mode = 'lines',
    name = 'Cumulative Real Error (m3)',
    line = dict(
        color = '#F44336',
        width = 3,
        dash = 'dot')
)

traceAB = go.Scatter(
    x = x,
    y = ab,
    mode = 'lines',
    name = 'Cumulative Absolute Error (m3)',
    line = dict(
        color = '#4527A0',
        width = 3,
        dash = 'dot')
)

layout = go.Layout(
    title='Water Balance Errors',
    xaxis=dict(
        title='',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title=r'$m^3$',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)
```

```python
data = [traceWB, traceRE, traceAB]
figname = "ErrorWater.png"

%matplotlib inline
import plotlycred

online = True
if online:
    # fig = py.iplot(data) 
    fig = go.Figure(data=data, layout=layout)
    py.image.save_as(fig, filename=figname)
    # py.plot(fig, filename=figname) # Opens in web
else:
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
```
