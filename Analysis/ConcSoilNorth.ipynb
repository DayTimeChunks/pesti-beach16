```python
import datetime
import os

import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import *
import plotly.graph_objs as go

import numpy as np

from obs_data import *

print(os.getcwd())
```

```python
init_notebook_mode(connected=True)
```

```python
PC = True
analysis = True
version = "model_v12"
folder = "1"
short_sim = True
```

```python
variable = "Smet Conc"
```

## Observed

### Composites

```python
n_obs = getSoilData("North", tss=True)
n_obs.head()
```

```python
if variable == "d13C":
    name = "comp.d13C"
else:
    name = 'Conc.mug.g.dry.soil'

nConc = n_obs[name].tolist()
```

### Detailed

```python
det_name = "ug.g"
detail_plots = ['N1', 'N2', 'N3', 'N4', 'N5', 'N7', 'N8']
detail_data = []

for i in range(len(detail_plots)):
    n = getDetailed(detail_plots[i] + ".tss")
    n = n[det_name].tolist()
    detail_data.append(n)
```

## Simulated

```python
if analysis:
    dot = ''
else:
    dot = '..\\'
    
if PC:
    back = "\\"  
    sim_path = dot + version + back  
else:
    back = "/"
    sim_path = version + back
    
print(sim_path)
```

```python
# Model version and mc-run number:
# version = "v10"
# folder = "1"

# Working computer
PC = False
if PC:
    back = "\\"  
    sim_path = version + back  
else:
    back = "/"
    sim_path = version + back
    
print(sim_path)
```

```python
# Time series name
col = 'sim' 
name_tss = 'CONC'
```

```python
transect = 'nor'
filename = "resM_" + transect + name_tss + ".tss"
c_north_sim = pd.read_table(sim_path + folder + back + filename, 
                            skiprows=4, delim_whitespace=True,
                            names=['Jdays', col],
                            header=None)

simN = c_north_sim[col].tolist()
```

## Plotting

```python
# Observations 
if short_sim:
    s_obs = 166 - 1
    e_obs = 300
else:
    # Julian days
    s_obs = 0
    e_obs = 300

# Julian days
x = n_obs.Jdays.tolist()

# Calendar days
dates = getTime(sep=";")

# x = dates.Date.tolist()
# q_obs[s_obs:(s_obs+20)]
```

### Observed

#### Composites

```python
nObs = go.Scatter(
    x = x[s_obs:e_obs],
    y = nConc[s_obs:e_obs],
    mode = 'markers',
    name = 'North '  + variable,
    marker=dict(
        color= 'red', # 'rgba(204, 204, 204, 0.95)',
        line=dict(
            color= 'black', #'rgba(217, 217, 217, 1.0)',
            width=1,
        ),
        symbol='circle',
        size=10)
)


```

#### Detailed

```python
data = []

for i in range(len(detail_plots)):
    n_det = go.Scatter(
        x = x[s_obs:e_obs],
        y = detail_data[i][s_obs:e_obs],
        mode = 'markers',
        name = detail_plots[i],
        marker=dict(
            color= 'red', # 'rgba(204, 204, 204, 0.95)',
            line=dict(
                color= 'black', #'rgba(217, 217, 217, 1.0)',
                width=1,
            ),
            symbol='star',
            size=10)
    )
    data.append(n_det)
```

### Simulated

```python
traceN = go.Scatter(
    x = x[s_obs:e_obs],
    y = simN,
    mode = 'lines',
    name = 'North Sim.',
    marker=dict(
        color= 'red')
)
```

```python
data.append(nObs)
data.append(traceN)
```

```python
%matplotlib inline
import plotlycred
figname = 'SoilsConc.png'

online = False
if online:
    # fig = py.iplot(data) 
    fig = go.Figure(data=data) # , layout=layout)
    py.image.save_as(fig, filename=figname)
else:
    iplot(data)
```
